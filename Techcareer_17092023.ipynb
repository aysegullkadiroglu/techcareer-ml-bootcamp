```python
vr.drop_duplicates()
```

```python
vr[vr.duplicated()==False]
```

```python
vr['c'] = range(6)
vr
```

```python
vr[vr.duplicated()==False]
```

```python
vr.duplicated(['a','b'])
```

```python
vr2 = pd.DataFrame({'isim':['Ali','caner','Nur','fatma'],
                    'notlar':[90,12,85,100]})
vr2
```

```python
vr2['şube'] = adlar.map(sınıflar)
vr2
```
