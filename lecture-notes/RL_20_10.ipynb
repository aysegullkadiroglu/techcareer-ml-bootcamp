```python
import pandas as pd
import numpy as np
import pylab as pl
!pip install networkx
import networkx as nx
```

```python
edge =[(0,1),(1,5),(5,6),(5,4),(1,2),
      (1,3),(9,10),(2,4),(0,6),(6,7),
      (8,9),(7,8),(1,7),(3,9)]

goal = 10
G = nx.Graph()
G.add_edges_from(edge)
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos) #düğüm
nx.draw_networkx_edges(G, pos) #bağlar
nx.draw_networkx_labels(G, pos) #etiket değerleri
pl.show()
```

```python
MATRIX_SIZE = 11
M = np.matrix(np.ones(shape=(MATRIX_SIZE, MATRIX_SIZE)))
M*= -1

for points in edge:
    print(points)
    if points[1]==goal:
        M[points]=100
    else:
        M[points]=0

    if points[0]== goal:
        M[points[::-1]]=100
    else:
        M[points[::-1]]=0
#M[goal, goal]=100
print(M)
```
